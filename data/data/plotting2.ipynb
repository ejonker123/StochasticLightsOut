```python
coeff = 0.05
for greedy_data in [stochastic_0_05, stochastic_0_10, stochastic_0_15,
                              stochastic_0_20, stochastic_0_25, stochastic_0_30]:
    max_value = greedy_data["step_count"].max()+200

    bins = np.arange(0, max_value + 1, 1)
    counts, bin_edges = np.histogram(greedy_data["step_count"], bins=bins)

    x_half = greedy_data["step_count"].quantile(0.5)
    threshold_95 = greedy_data["step_count"].quantile(0.95)

    sns.histplot(
        greedy_data["step_count"],
        bins=range(0, max_value, int(1 / (50 / max_value)) if max_value >= 50 else 1),
        kde=True,
        label="step_count",
        color="orange",
        edgecolor="black"
    )
    print("Mean:", round(greedy_data["step_count"].mean(), 2))
    plt.title(f"Stochastic Coefficient {coeff:0.2f} 3b3")
    plt.axvline(threshold_95, color="green", linestyle="--", linewidth=2, label=f"95% threshold = {int(threshold_95)}")

    plt.axvline(x_half, color="red", linestyle="--", linewidth=2, label=f"median = {int(x_half)}")
    plt.xlabel("step_count")
    plt.ylabel("Frequency")
    plt.grid(axis="y", alpha=0.8)
    plt.legend()
    plt.savefig(f"stochastic_0_{int(coeff * 100)}_3b3.png")
    plt.show()

    coeff += 0.05
```
